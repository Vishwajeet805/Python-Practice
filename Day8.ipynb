```python
# Empty dictionary
dog = {}

# Add values
dog["name"] = "Buddy"
dog["color"] = "Brown"
dog["breed"] = "Labrador"
dog["legs"] = 4
dog["age"] = 3

print(dog)

# Student dictionary
student = {
    "first_name": "Shivam",
    "last_name": "Singh",
    "gender": "Male",
    "age": 24,
    "marital_status": "Single",
    "skills": ["Python", "SQL"],
    "country": "India",
    "city": "Jabalpur",
    "address": "MP"
}

print(student)
```

```python
# Length
print(len(student))

# Skills and type
print(student["skills"])
print(type(student["skills"]))

# Modify skills
student["skills"].append("Data Analysis")
student["skills"].append("Machine Learning")

print(student["skills"])

# Keys and values
print(list(student.keys()))
print(list(student.values()))

# Items to list of tuples
print(list(student.items()))
```

```python
# Delete one item
student.pop("address")

print(student)

# Delete dictionary
del student
```
