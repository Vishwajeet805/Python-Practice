```python
import random
import string

# Random user id
def random_user_id():
    chars = string.ascii_letters + string.digits
    return ''.join(random.choice(chars) for _ in range(6))

print(random_user_id()) 

# User-defined ID generator
def user_id_gen_by_user():
    length = int(input("Enter number of characters: "))
    count = int(input("Enter number of IDs: "))

    chars = string.ascii_letters + string.digits

    for _ in range(count):
        user_id = ''.join(random.choice(chars) for _ in range(length))
        print(user_id)


# RGB color generator
def rgb_color_gen():
    r = random.randint(0, 255)
    g = random.randint(0, 255)
    b = random.randint(0, 255)
    return f"rgb({r},{g},{b})"

print(rgb_color_gen())
```

```python
# Hexa colors
def list_of_hexa_colors(n):
    hex_chars = '0123456789abcdef'
    colors = []
    for _ in range(n):
        color = '#' + ''.join(random.choice(hex_chars) for _ in range(6))
        colors.append(color)
    return colors


# RGB colors list
def list_of_rgb_colors(n):
    colors = []
    for _ in range(n):
        colors.append(rgb_color_gen())
    return colors


# Generate colors
def generate_colors(color_type, n):
    if color_type == 'hexa':
        return list_of_hexa_colors(n)
    elif color_type == 'rgb':
        return list_of_rgb_colors(n)
    
generate_colors('hexa', 3)  
generate_colors('hexa', 1) 
generate_colors('rgb', 3)  
generate_colors('rgb', 1)  
```

```python
# Shuffle list
def shuffle_list(lst):
    random.shuffle(lst)
    return lst


# Unique random numbers
def unique_random_numbers():
    return random.sample(range(10), 7)
```
