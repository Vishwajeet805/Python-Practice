```python
# String concatenation
s1 = "Thirty"
s2 = "Days"
s3 = "Of"
s4 = "Python"

result = s1 + " " + s2 + " " + s3 + " " + s4
print(result)

coding = "Coding" + " " + "For" + " " + "All"
print(coding)

# Variable
company = "Coding For All"
print(company)

# Length
print(len(company))

# Case conversion
print(company.upper())
print(company.lower())

# Formatting
print(company.capitalize())
print(company.title())
print(company.swapcase())
```
