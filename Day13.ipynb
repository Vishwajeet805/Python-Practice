```python
# Filter negative and zero
numbers = [-4, -3, -2, -1, 0, 2, 4, 6]

filtered = [n for n in numbers if n <= 0]
print(filtered)


# Flatten list of lists
list_of_lists = [[1,2,3],[4,5,6],[7,8,9]]

flat = [num for sublist in list_of_lists for num in sublist]
print(flat)
```

```python
# List of tuples
result = [(i, 1, i, i**2, i**3, i**4, i**5) for i in range(11)]
print(result)


# Flatten country data
countries = [[('Finland', 'Helsinki')], [('Sweden', 'Stockholm')], [('Norway', 'Oslo')]]

flat_countries = [[c[0].upper(), c[0][:3].upper(), c[1].upper()] 
                  for sub in countries for c in sub]

print(flat_countries)


# Convert to dictionary
dict_countries = [{"country": c[0].upper(), "city": c[1].upper()} 
                  for sub in countries for c in sub]

print(dict_countries)
```

```python
# Concatenate names
names = [[('Asabeneh', 'Yetayeh')], [('David', 'Smith')], 
         [('Donald', 'Trump')], [('Bill', 'Gates')]]

full_names = [f"{n[0]} {n[1]}" for sub in names for n in sub]
print(full_names)


# Lambda functions
slope = lambda x1, y1, x2, y2: (y2 - y1) / (x2 - x1)
y_intercept = lambda x, y, m: y - m * x

print("Slope:", slope(2, 2, 6, 10))
print("Y-intercept:", y_intercept(2, 2, 2))
```
