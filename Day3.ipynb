```python
# Variables
age = 24
height = 5.9
complex_num = 2 + 3j

print(age, height, complex_num)


# Triangle Area
base = float(input("Enter base: "))
height = float(input("Enter height: "))

area = 0.5 * base * height
print("The area of the triangle is", area)


# Triangle Perimeter
a = float(input("Enter side a: "))
b = float(input("Enter side b: "))
c = float(input("Enter side c: "))

perimeter = a + b + c
print("The perimeter of the triangle is", perimeter)
```

```python
# Rectangle
length = float(input("Enter length: "))
width = float(input("Enter width: "))

area_rect = length * width
perimeter_rect = 2 * (length + width)

print("Area:", area_rect)
print("Perimeter:", perimeter_rect)


# Circle
radius = float(input("Enter radius: "))
pi = 3.14

area_circle = pi * radius * radius
circumference = 2 * pi * radius

print("Area:", area_circle)
print("Circumference:", circumference)


# Linear equation y = 2x - 2
# slope = 2
slope = 2
x_intercept = 1
y_intercept = -2

print("Slope:", slope)
print("x-intercept:", x_intercept)
print("y-intercept:", y_intercept)


# Slope and distance between points
import math

x1, y1 = 2, 2
x2, y2 = 6, 10

slope_2 = (y2 - y1) / (x2 - x1)
distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

print("Slope:", slope_2)
print("Distance:", distance)
```

```python
# Quadratic equation
for x in range(-5, 5):
    y = x**2 + 6*x + 9
    if y == 0:
        print("y is zero at x =", x)


# String checks
print(len("python") != len("dragon"))
print("on" in "python" and "on" in "dragon")
print("jargon" in "I hope this course is not full of jargon")
print("on" not in "dragon" and "on" not in "python")


# Type conversions
length = len("python")
print(float(length))
print(str(length))


# Even check
num = int(input("Enter number: "))
print(num % 2 == 0)


# Floor division check
print(7 // 3 == int(2.7))


# Type comparison
print(type("10") == type(10))


# Conversion check
try:
    print(int("9.8") == 10)
except:
    print("Conversion not possible")


# Weekly earning
hours = float(input("Enter hours: "))
rate = float(input("Enter rate per hour: "))

pay = hours * rate
print("Your weekly earning is", pay)


# Seconds lived
years = int(input("Enter number of years: "))
seconds = years * 365 * 24 * 60 * 60

print("You have lived for", seconds, "seconds")


# Table
for i in range(1, 6):
    print(i, 1, i, i**2, i**3)
```
