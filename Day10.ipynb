```python
# For loop 0 to 10
for i in range(11):
    print(i)

# While loop 0 to 10
i = 0
while i <= 10:
    print(i)
    i += 1

# For loop 10 to 0
for i in range(10, -1, -1):
    print(i)

# While loop 10 to 0
i = 10
while i >= 0:
    print(i)
    i -= 1

# Triangle pattern
for i in range(1, 8):
    print("#" * i)

# Square pattern
for i in range(8):
    for j in range(8):
        print("#", end=" ")
    print()

# Multiplication squares
for i in range(11):
    print(f"{i} x {i} = {i*i}")
```

```python
# Iterate list
items = ['Python', 'Numpy', 'Pandas', 'Django', 'Flask']

for item in items:
    print(item)

# Even numbers
for i in range(101):
    if i % 2 == 0:
        print(i)

# Odd numbers
for i in range(101):
    if i % 2 != 0:
        print(i)
```

```python
# Sum of all numbers
total = 0
for i in range(101):
    total += i
print("The sum of all numbers is", total)

# Sum of evens and odds
even_sum = 0
odd_sum = 0

for i in range(101):
    if i % 2 == 0:
        even_sum += i
    else:
        odd_sum += i

print("The sum of all evens is", even_sum)
print("The sum of all odds is", odd_sum)


# Countries containing 'land'
countries = ['Finland', 'Iceland', 'Thailand', 'India', 'England']

for country in countries:
    if 'land' in country:
        print(country)


# Reverse fruits
fruits = ['banana', 'orange', 'mango', 'lemon']

for i in range(len(fruits)-1, -1, -1):
    print(fruits[i])


# Languages count (sample data)
countries_data = [
    {"name": "India", "languages": ["Hindi", "English"], "population": 1400000000},
    {"name": "USA", "languages": ["English"], "population": 330000000},
    {"name": "Finland", "languages": ["Finnish", "Swedish"], "population": 5500000}
]

# Total languages
all_languages = set()
for country in countries_data:
    for lang in country["languages"]:
        all_languages.add(lang)

print("Total languages:", len(all_languages))

# Most spoken languages
lang_count = {}
for country in countries_data:
    for lang in country["languages"]:
        lang_count[lang] = lang_count.get(lang, 0) + 1

sorted_langs = sorted(lang_count.items(), key=lambda x: x[1], reverse=True)
print("Most spoken languages:", sorted_langs[:10])

# Most populated countries
sorted_pop = sorted(countries_data, key=lambda x: x["population"], reverse=True)
print("Most populated countries:", sorted_pop[:10])
```
