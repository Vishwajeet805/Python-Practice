```python
# For loop 0 to 10
for i in range(11):
    print(i)

# While loop 0 to 10
i = 0
while i <= 10:
    print(i)
    i += 1

# For loop 10 to 0
for i in range(10, -1, -1):
    print(i)

# While loop 10 to 0
i = 10
while i >= 0:
    print(i)
    i -= 1

# Triangle pattern
for i in range(1, 8):
    print("#" * i)

# Square pattern
for i in range(8):
    for j in range(8):
        print("#", end=" ")
    print()

# Multiplication squares
for i in range(11):
    print(f"{i} x {i} = {i*i}")
```
