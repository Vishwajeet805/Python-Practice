```python
# Check Python version
import sys
print("Python Version:", sys.version)

# Arithmetic Operations
a = 3
b = 4

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Modulus:", a % b)
print("Division:", a / b)
print("Exponential:", a ** b)
print("Floor Division:", a // b)
```

```python
# Strings
print("Your Name")
print("Your Family Name")
print("India")
print("I am enjoying 30 days of python")

# Data Types
print(type(10))
print(type(9.8))
print(type(3.14))
print(type(4 - 4j))
print(type(['Asabeneh', 'Python', 'Finland']))
print(type("Your Name"))
print(type("Your Family Name"))
print(type("India"))
```

```python
# Different Data Types Examples

# Number
integer_num = 10
float_num = 3.14
complex_num = 2 + 3j

# String
string_text = "Hello Python"

# Boolean
is_learning = True

# List
my_list = [1, 2, 3, "Python"]

# Tuple
my_tuple = (1, 2, 3)

# Set
my_set = {1, 2, 3}

# Dictionary
my_dict = {"name": "Shivam", "country": "India"}

print(type(integer_num))
print(type(float_num))
print(type(complex_num))
print(type(string_text))
print(type(is_learning))
print(type(my_list))
print(type(my_tuple))
print(type(my_set))
print(type(my_dict))


# Euclidean Distance between (2,3) and (10,8)
import math

x1, y1 = 2, 3
x2, y2 = 10, 8

distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

print("Euclidean Distance:", distance)
```
