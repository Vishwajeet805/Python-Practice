```python
# Empty tuple
empty_tuple = ()

# Siblings
brothers = ("Rahul", "Aman")
sisters = ("Riya", "Neha")

# Join tuples
siblings = brothers + sisters

# Count siblings
print("Siblings:", siblings)
print("Total siblings:", len(siblings))

# Add parents
family_members = siblings + ("Father", "Mother")
print("Family members:", family_members)
```

```python
# Unpack siblings and parents
*siblings_only, father, mother = family_members
print("Siblings:", siblings_only)
print("Parents:", father, mother)

# Food tuples
fruits = ("Apple", "Banana", "Mango")
vegetables = ("Carrot", "Spinach", "Potato")
animal_products = ("Milk", "Egg", "Meat")

# Join tuples
food_stuff_tp = fruits + vegetables + animal_products
print(food_stuff_tp)

# Convert to list
food_stuff_lt = list(food_stuff_tp)
print(food_stuff_lt)

# Middle item
mid = len(food_stuff_lt)//2
print("Middle item:", food_stuff_lt[mid])

# First and last 3 items
print("First 3:", food_stuff_lt[:3])
print("Last 3:", food_stuff_lt[-3:])
```

```python
# Delete tuple
del food_stuff_tp

# Nordic countries check
nordic_countries = ('Denmark', 'Finland', 'Iceland', 'Norway', 'Sweden')

print("Estonia" in nordic_countries)
print("Iceland" in nordic_countries)
```
