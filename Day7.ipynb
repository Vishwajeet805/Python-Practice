```python
# Sets
it_companies = {'Facebook', 'Google', 'Microsoft', 'Apple', 'IBM', 'Oracle', 'Amazon'}

# Length
print(len(it_companies))

# Add item
it_companies.add('Twitter')

# Add multiple items
it_companies.update(['Tesla', 'Netflix'])

# Remove item
it_companies.remove('Oracle')

# Difference between remove and discard
# remove() → error if item not found
# discard() → no error if item not found
```

```python
A = {19, 22, 24, 20, 25, 26}
B = {19, 22, 20, 25, 26, 24, 28, 27}

# Union
print(A.union(B))

# Intersection
print(A.intersection(B))

# Subset
print(A.issubset(B))

# Disjoint
print(A.isdisjoint(B))

# Symmetric difference
print(A.symmetric_difference(B))

# Delete sets
del A
del B
```

```python
# Ages
age = [22, 19, 24, 25, 26, 24, 25, 24]

# Convert to set
age_set = set(age)

print("List length:", len(age))
print("Set length:", len(age_set))

# Difference explanation
# string → ordered, immutable
# list → ordered, mutable
# tuple → ordered, immutable
# set → unordered, unique elements

# Unique words
sentence = "I am a teacher and I love to inspire and teach people"

words = sentence.split()
unique_words = set(words)

print("Unique words:", unique_words)
print("Count:", len(unique_words))
```
