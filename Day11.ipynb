```python
import math

# Add two numbers
def add_two_numbers(a, b):
    return a + b

# Area of circle
def area_of_circle(r):
    return math.pi * r * r

# Add all nums
def add_all_nums(*args):
    total = 0
    for i in args:
        if not isinstance(i, (int, float)):
            return "All inputs must be numbers"
        total += i
    return total

# Celsius to Fahrenheit
def convert_celsius_to_fahrenheit(c):
    return (c * 9/5) + 32

# Check season
def check_season(month):
    if month in ["September", "October", "November"]:
        return "Autumn"
    elif month in ["December", "January", "February"]:
        return "Winter"
    elif month in ["March", "April", "May"]:
        return "Spring"
    elif month in ["June", "July", "August"]:
        return "Summer"

# Slope
def calculate_slope(x1, y1, x2, y2):
    return (y2 - y1) / (x2 - x1)

# Quadratic equation
def solve_quadratic_eqn(a, b, c):
    d = b**2 - 4*a*c
    if d < 0:
        return "No real solution"
    x1 = (-b + math.sqrt(d)) / (2*a)
    x2 = (-b - math.sqrt(d)) / (2*a)
    return x1, x2
```

```python
# Print list
def print_list(lst):
    for item in lst:
        print(item)

# Reverse list
def reverse_list(arr):
    rev = []
    for i in range(len(arr)-1, -1, -1):
        rev.append(arr[i])
    return rev

# Capitalize list items
def capitalize_list_items(lst):
    return [item.upper() for item in lst]

# Add item
def add_item(lst, item):
    lst.append(item)
    return lst

# Remove item
def remove_item(lst, item):
    if item in lst:
        lst.remove(item)
    return lst

# Sum of numbers
def sum_of_numbers(n):
    return sum(range(n+1))

# Sum of odds
def sum_of_odds(n):
    return sum(i for i in range(n+1) if i % 2 != 0)

# Sum of evens
def sum_of_even(n):
    return sum(i for i in range(n+1) if i % 2 == 0)

# Evens and odds count
def evens_and_odds(n):
    evens = sum(1 for i in range(n+1) if i % 2 == 0)
    odds = sum(1 for i in range(n+1) if i % 2 != 0)
    return f"The number of odds are {odds}. The number of evens are {evens}."

# Factorial
def factorial(n):
    result = 1
    for i in range(1, n+1):
        result *= i
    return result

# Check empty
def is_empty(val):
    return not bool(val)
```

```python
import statistics

# Stats functions
def calculate_mean(lst):
    return sum(lst)/len(lst)

def calculate_median(lst):
    return statistics.median(lst)

def calculate_mode(lst):
    return statistics.mode(lst)

def calculate_range(lst):
    return max(lst) - min(lst)

def calculate_variance(lst):
    return statistics.variance(lst)

def calculate_std(lst):
    return statistics.stdev(lst)

# Greet
def greet(name="Guest"):
    return f"Hello, {name}!"

# Named args
def show_args(**kwargs):
    return ", ".join([f"{k}: {v}" for k, v in kwargs.items()])

# Prime check
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

# Unique items
def all_unique(lst):
    return len(lst) == len(set(lst))

# Same type
def same_type(lst):
    return len(set(type(i) for i in lst)) == 1

# Valid variable
def is_valid_var(name):
    return name.isidentifier()

# Most spoken languages (sample)
def most_spoken_languages(data, n=10):
    lang_count = {}
    for country in data:
        for lang in country["languages"]:
            lang_count[lang] = lang_count.get(lang, 0) + 1
    return sorted(lang_count.items(), key=lambda x: x[1], reverse=True)[:n]

# Most populated countries (sample)
def most_populated_countries(data, n=10):
    return sorted(data, key=lambda x: x["population"], reverse=True)[:n]
```
