```python
# Empty list
empty_list = []

# List with more than 5 items
my_list = [1, 2, 3, 4, 5, 6]

# Length
print(len(my_list))

# First, middle, last
print(my_list[0])
print(my_list[len(my_list)//2])
print(my_list[-1])

# Mixed data types
mixed_data_types = ["Shivam", 24, 5.9, False, "India"]

# IT companies
it_companies = ["Facebook", "Google", "Microsoft", "Apple", "IBM", "Oracle", "Amazon"]

print(it_companies)
print(len(it_companies))

# First, middle, last company
print(it_companies[0])
print(it_companies[len(it_companies)//2])
print(it_companies[-1])
```

```python
# Modify one company
it_companies[1] = "Meta"

# Add company
it_companies.append("Tesla")

# Insert in middle
it_companies.insert(len(it_companies)//2, "Infosys")

# Uppercase one (except IBM)
it_companies[0] = it_companies[0].upper()

# Join
print("#; ".join(it_companies))

# Check existence
print("Google" in it_companies)

# Sort and reverse
it_companies.sort()
it_companies.reverse()

# Slice
print(it_companies[:3])
print(it_companies[-3:])
print(it_companies[len(it_companies)//2])

# Remove operations
it_companies.pop(0)
it_companies.pop(len(it_companies)//2)
it_companies.pop(-1)

# Clear and delete
it_companies.clear()
del it_companies
```

```python
# Joining lists
front_end = ['HTML', 'CSS', 'JS', 'React', 'Redux']
back_end = ['Node','Express', 'MongoDB']

full_stack = front_end + back_end
full_stack.insert(full_stack.index("Redux") + 1, "Python")
full_stack.insert(full_stack.index("Python") + 1, "SQL")

print(full_stack)


# Ages
ages = [19, 22, 19, 24, 20, 25, 26, 24, 25, 24]

ages.sort()
print("Min:", min(ages))
print("Max:", max(ages))

ages.append(min(ages))
ages.append(max(ages))

# Median
ages.sort()
mid = len(ages)//2
median = (ages[mid] + ages[mid-1]) / 2
print("Median:", median)

# Average
avg = sum(ages)/len(ages)
print("Average:", avg)

# Range
print("Range:", max(ages) - min(ages))

# Compare
print(abs(min(ages) - avg), abs(max(ages) - avg))


# Countries
countries = ['China', 'Russia', 'USA', 'Finland', 'Sweden', 'Norway', 'Denmark']

mid = len(countries)//2
print(countries[mid])

# Split list
first_half = countries[:mid+1]
second_half = countries[mid+1:]

print(first_half, second_half)

# Unpack
china, russia, usa, *scandic = countries
print(china, russia, usa, scandic)
```
